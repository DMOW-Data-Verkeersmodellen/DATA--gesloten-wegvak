### Merge links - count obv link-no

```python
links = links.merge(
    count[['LINKNO', 'FROMNODENO', 'TONODENO', 'NO', 'CODE', 'SG_ID', 'TEL_PW_ETM', 'TEL_VR_ETM']],
    left_on=['NO', 'FROMNODENO', 'TONODENO'],
    right_on=['LINKNO', 'FROMNODENO', 'TONODENO'],
    how='left',
    suffixes=('', '_count')
)
```

### Algoritme

```python
links = links.rename(columns={'FROMNODENO': 'from_node', 'TONODENO': 'to_node'}) # zo in gpt
links.columns
```

```python
import networkx as nx
G = nx.DiGraph()  # Gerichte graaf voor het netwerk:contentReference[oaicite:3]{index=3}.
for idx, row in links.iterrows():
    u = row['from_node']
    v = row['to_node']
    G.add_node(u)
    G.add_node(v)
    # Voeg de link (boog) toe met attributen: type en index
    G.add_edge(u, v, id=idx, type=row['VISUEELL~1'])
```
